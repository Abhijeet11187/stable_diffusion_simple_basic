```python
# 1. Stable Diffusion is an AI model that creates images from text prompts like “cat wearing sunglasses.”

# 2. It starts with random noise (like TV static) and slowly removes noise to form a clear image.

# 3. It uses deep learning models such as CLIP, UNet, and VAE to understand text and generate images.

# 4. Stable Diffusion works in a compressed hidden space called “latent space,” making it faster and cheaper than older diffusion models.

# 5. It is open-source, so developers can run, fine-tune, and customize it on their own systems.

# 6. It is widely used for AI art, YouTube thumbnails, game assets, image editing, anime generation, and design concepts.

```

```python
from diffusers import DiffusionPipeline
```

```python
pipeline=DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5",USE_SAFETENSORS=True)
```
